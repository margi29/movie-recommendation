```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

# Function to extract keywords from the overview using TF-IDF
def extract_keywords_from_overview_tfidf(overview):
    tfidf_vectorizer = TfidfVectorizer(stop_words='english')
    tfidf_matrix = tfidf_vectorizer.fit_transform([overview])
    feature_names = tfidf_vectorizer.get_feature_names_out()
    # Sort the features by their TF-IDF scores and select the top N keywords
    top_n = 10
    top_keywords_indices = tfidf_matrix.toarray().argsort()[0][-top_n:][::-1]
    top_keywords = [feature_names[idx] for idx in top_keywords_indices]
    return top_keywords

# Load your DataFrame (assuming it's named 'df')
# Loop through each row and extract keywords from overview using TF-IDF
added_keywords_per_movie = {}
for index, row in df.iterrows():
    if pd.isnull(row['keywords']):  # Only process rows with NaN keywords
        keywords_from_overview = extract_keywords_from_overview_tfidf(row['overview'])
        # Join the extracted keywords and update the 'keywords' column
        if keywords_from_overview:
            df.at[index, 'keywords'] = ', '.join(keywords_from_overview)
            added_keywords_per_movie[row['title']] = keywords_from_overview

# Print the added keywords for each movie
for movie_title, keywords_added in added_keywords_per_movie.items():
    print(f"Movie: {movie_title}")
    for keyword_added in keywords_added:
        print(f"    Added Keyword: {keyword_added}")
```

```python
df['tags'] = df['overview']+df['genres']+df['keywords']+df['casts']+df['directors']
```

```python
# Convert non-string values to empty string
df['tags'] = df['tags'].apply(lambda x: '' if not isinstance(x, str) else x)

# Convert tags to lowercase
df['tags'] = df['tags'].apply(lambda x: x.lower())
```

```python
df.head(2)
```

```python
def preprocess_dataset(df):
    # Convert runtime to H and min format
    df['runtime'] = df['runtime'].apply(lambda x: f"{x//60}H {x%60}min")

    # Convert release date to date format
    df['release_date'] = pd.to_datetime(df['release_date'])
    
    df['vote_average'] = df['vote_average'].round(1)
    
    df['popularity'] = df['popularity'].round(2)
    
    df = df.sort_values('release_date', ascending=False)
    return df



# Preprocess the dataset
new = preprocess_dataset(df)

# Save the preprocessed dataset
new.to_csv('dataset/preprocessed_movies.csv', index=False)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
from sklearn.neighbors import NearestNeighbors

def recommend(movie, k=10):
    # Convert the search query to lowercase
    movie = movie.lower()
    
    # Find the index of the movie in the DataFrame
    index = new[new['title'].str.lower() == movie].index
    
    if len(index) == 0:
        print("Movie not found.")
        return
    
    index = index[0]
    
    # Initialize the KNN model
    knn = NearestNeighbors(n_neighbors=k, metric='cosine')
    knn.fit(similarity)  # similarity is the similarity matrix
    
    # Get the indices of the nearest neighbors
    distances, indices = knn.kneighbors([similarity[index]])
    
    # Print the titles of the top k similar movies
    for i in indices[0][1:]:
        print(new.iloc[i].title)

```
