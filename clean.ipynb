```python
def process_genres(genres_str):
    # Split the string by comma and strip whitespace
    genres_list = genres_str.split(',')
    return genres_list

# Apply the function to the 'genres' column
movies['genres'] = movies['genres'].apply(lambda x: process_genres(x) if isinstance(x, str) else [])
```

```python
movies.iloc[0].genres
```

```python
movies.head(1)
```

```python
def process_casts(casts_str):
    # Split the string by comma and strip whitespace
    casts_list = casts_str.split(',')
    # Take the first four members if available, otherwise return an empty list
    first_four_casts = casts_list[:4]
    # Strip whitespace from each cast member's name
    first_four_casts = [cast.strip() for cast in first_four_casts]
    return first_four_casts

# Apply the function to the 'casts' column
movies['casts'] = movies['casts'].apply(lambda x: process_casts(x) if isinstance(x, str) else [])
```

```python
movies.drop('crew', axis=1, inplace=True)
```

```python
# Apply the function to the entire 'keywords' column
movies['keywords'] = movies['keywords'].apply(lambda x: x.split(', '))
movies['directors'] = movies['directors'].apply(lambda x: x.split(', '))
# Check for NaN values in the 'videos' column
nan_indices = movies['videos'].isnull()

# Apply splitting operation only to non-null values
movies.loc[~nan_indices, 'videos'] = movies.loc[~nan_indices, 'videos'].apply(lambda x: x.split(', '))
```

```python
movies.head(3)
```

```python
def collapse(obj):
    l = []
    for i in obj:
        l.append(i.replace(" ",""))
    return l
```

```python
movies['casts'] = movies['casts'].apply(collapse)
movies['directors'] = movies['directors'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head(1)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags'] = movies['overview']+movies['genres']+movies['keywords']+movies['casts']+movies['directors']
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y) 
```

```python
new['tags'] = new['tags'].apply(stem)
```

```python
new.info()
```

```python
import os
import pandas as pd

# Assuming you have DataFrame `new` already created

# Define the path to your dataset folder
dataset_folder = 'dataset'

# Create the dataset folder if it doesn't exist
if not os.path.exists(dataset_folder):
    os.makedirs(dataset_folder)

# Define the file path for saving the CSV
csv_filename = os.path.join(dataset_folder, 'new_data.csv')

# Check if the file already exists
if not os.path.exists(csv_filename):
    # Save the DataFrame to CSV
    new.to_csv(csv_filename, index=False)
    print(f"DataFrame saved to {csv_filename}")
else:
    print(f"The file {csv_filename} already exists. Data was not overwritten.")
```

```python
#countvectorization
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features = 5000, stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
#recommend
def recommend(movie):
    # Convert the search query to lowercase
    movie = movie.lower()
    
    # Find the index of the movie in the DataFrame
    index = new[new['title'].str.lower() == movie].index[0]
    
    # Sort distances based on similarity
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])
    
    # Print the titles of the top 5 similar movies
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```
